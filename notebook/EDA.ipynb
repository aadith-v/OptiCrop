# Crop Recommendation Dataset Analysis

## Univariate Analysis


```python
plt.figure(figsize=(12,6))
sns.countplot(x='label', data=df)
plt.xticks(rotation=90)
plt.title("Crop Distribution")
plt.show()
```

### Observation

The crop distribution is balanced, with each crop having approximately the same number of samples. This balanced distribution is beneficial for training machine learning models, as it reduces the risk of bias toward any particular crop.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['temperature'], bins=20, kde=True)
plt.title("Temperature Distribution")
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```

### Observation

The temperature values range from approximately 8°C to 43°C, with most samples concentrated between 20°C and 30°C. The highest frequency is around 25°C–28°C, indicating that most crops in the dataset grow under moderate temperature conditions.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['humidity'], bins=20, kde=True, color='green')
plt.title("Humidity Distribution")
plt.xlabel("Humidity")
plt.ylabel("Frequency")
plt.show()
```

### Observation

The humidity values range from low to high levels, with most observations concentrated in the middle to higher humidity range. This indicates that the dataset contains crops suitable for different humidity conditions.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['ph'], bins=20, kde=True, color='orange')
plt.title("pH Distribution")
plt.xlabel("pH")
plt.ylabel("Frequency")
plt.show()
```

### Observation

Most soil pH values are between 5.5 and 7.5, indicating that the majority of crops prefer slightly acidic to neutral soil conditions.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['rainfall'], bins=20, kde=True, color='blue')
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall")
plt.ylabel("Frequency")
plt.show()
```

### Observation

Rainfall values vary over a wide range, showing that the dataset includes crops suitable for different rainfall conditions.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['N'], bins=20, kde=True, color='purple')
plt.title("Nitrogen (N) Distribution")
plt.xlabel("Nitrogen")
plt.ylabel("Frequency")
plt.show()
```

### Observation

Nitrogen values are spread across different levels, indicating variation in soil nutrient content among the crop samples.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['P'], bins=20, kde=True, color='red')
plt.title("Phosphorous (P) Distribution")
plt.xlabel("Phosphorous")
plt.ylabel("Frequency")
plt.show()
```

### Observation

The phosphorous values show a broad distribution, suggesting that different crops require different phosphorous levels.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['K'], bins=20, kde=True, color='brown')
plt.title("Potassium (K) Distribution")
plt.xlabel("Potassium")
plt.ylabel("Frequency")
plt.show()
```

### Observation

Potassium values vary across the dataset, reflecting different potassium requirements for various crop types.

```python
plt.figure(figsize=(15,6))
sns.boxplot(x='label', y='temperature', data=df)
plt.xticks(rotation=90)
plt.title("Temperature vs Crop")
plt.show()
```

### Observation

Different crops grow under different temperature ranges. Some crops prefer lower temperatures, while others require warmer climatic conditions.

## Multivariate Analysis

```python
plt.figure(figsize=(10,8))
sns.heatmap(correlation, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()
```

### Observation

The heatmap shows the correlation between all numerical features. Most features have weak to moderate correlations, indicating that each feature contributes unique information for crop prediction.

```python
sns.pairplot(df[['N','P','K','temperature','humidity','ph','rainfall']])
plt.show()
```

### Observation
